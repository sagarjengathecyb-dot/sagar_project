```python
# find  yearly and monthly trend

# Extract year and month from order date
dataset['Year'] = dataset['Order Date'].dt.year
dataset['Month'] = dataset['Order Date'].dt.month

# Aggregate sales by year and month
monthly_sales = dataset.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

# Create a datetime column for plotting
monthly_sales['YearMonth'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(DAY=1))

# Plot monthly sales trend
plt.figure(figsize=(18,6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Time')
plt.ylabel('Total Sales (₹)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# import statsmodels holt-winters for smoothing model for time siries forecasting
# create aggreate to convers date time to year -month
# Converts period index back to proper datetime format
#for Time series models require datetime index, not period index


from statsmodels.tsa.holtwinters import ExponentialSmoothing
# Aggregate monthly sales
dataset['YearMonth'] = dataset['Order Date'].dt.to_period('M')
monthly_sales = dataset.groupby('YearMonth')['Sales'].sum()

# Convert to datetime index for time series model
monthly_sales.index = monthly_sales.index.to_timestamp()

# Holt-Winters Exponential Smoothing model
model = ExponentialSmoothing(monthly_sales, seasonal='mul', seasonal_periods=12).fit()

# Forecast next 12 months
forecast = model.forecast(12)

# Plot actual sales and forecast
plt.figure(figsize=(12,6))
plt.plot(monthly_sales, label='Actual Sales')
plt.plot(forecast, label='Forecasted Sales', linestyle='--')
plt.title('Monthly Sales Forecast Using Holt-Winters')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
