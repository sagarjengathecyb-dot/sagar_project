```python
# Only select numeric columns of interest
corr_matrix = dataset[['Sales', 'Profit', 'Discount']].corr()
print("Correlation Matrix:")
print(corr_matrix)

# Visualizing the correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Sales, Profit, and Discount')
plt.show()
```

```python
# Feature Importance using Random Forest
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Convert customer name to numeric code for modeling
le = LabelEncoder()
dataset['Customer_Code'] = le.fit_transform(dataset['Customer Name'])
model_data = dataset[['Sales', 'Profit', 'Discount', 'Customer_Code']].dropna()

X = model_data[['Profit', 'Discount', 'Customer_Code']]
y = model_data['Sales']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Fit Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=1)
rf.fit(X_train, y_train)

# Display feature importances
importances = rf.feature_importances_
feat_names = X.columns
for feat, importance in zip(feat_names, importances):
    print(f"Feature: {feat}, Importance: {importance:.4f}")
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans


# Prepare feature set for clustering
# Example: Segmenting customers based on their total sales, total profit, and total discount

customer_agg = dataset.groupby('Customer Name').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Discount': 'mean'  # or sum, as appropriate
}).reset_index()

# Scale features for K-Means
scaler = StandardScaler()
features_scaled = scaler.fit_transform(customer_agg[['Sales', 'Profit', 'Discount']])

# Determine appropriate number of clusters (elbow method, not shown here)
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
customer_agg['Cluster'] = kmeans.fit_predict(features_scaled)

# You can inspect cluster means:
print(customer_agg.groupby('Cluster')[['Sales', 'Profit', 'Discount']].mean())

# To see which cluster a customer is in:
print(customer_agg[['Customer Name', 'Cluster']].head())

# Optional: Visualize clusters
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(customer_agg['Sales'], customer_agg['Profit'], c=customer_agg['Cluster'], cmap='viridis')
plt.xlabel('Total Sales')
plt.ylabel('Total Profit')
plt.title('Customer Segmentation by Sales & Profit Clusters')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Aggregate Sales by month
monthly_sales = dataset.set_index('Order Date').resample('ME')['Sales'].sum()

# Plot the time series of monthly sales
plt.figure(figsize=(12,6))
plt.plot(monthly_sales)
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()

# Seasonal decomposition to extract trend, seasonality, and residuals
result = seasonal_decompose(monthly_sales, model='additive')
result.plot()
plt.show()
```

```python


# Aggregate KPIs by Region
region_kpi = dataset.groupby('Region').agg({
    'Sales': ['sum', 'mean', 'std'],
    'Profit': ['sum', 'mean', 'std'],
    'Discount': ['mean']
}).reset_index()
region_kpi.columns = ['Region', 'Total Sales', 'Avg Sales', 'Sales Std', 'Total Profit', 'Avg Profit', 'Profit Std', 'Avg Discount']

print(region_kpi)

# Visualize average sales and profit by region
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Avg Sales', data=region_kpi)
plt.title('Average Sales by Region')
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Avg Profit', data=region_kpi)
plt.title('Average Profit by Region')
plt.show()

# Outlier detection - using Z-score for Sales and Profit
from scipy.stats import zscore

dataset['Sales_zscore'] = zscore(dataset['Sales'])
dataset['Profit_zscore'] = zscore(dataset['Profit'])

# Define outliers as abs(zscore) > 3
sales_outliers = dataset[np.abs(dataset['Sales_zscore']) > 3]
profit_outliers = dataset[np.abs(dataset['Profit_zscore']) > 3]

print("Sales Outliers:")
print(sales_outliers[['Order ID', 'Region', 'Sales', 'Sales_zscore']])

print("Profit Outliers:")
print(profit_outliers[['Order ID', 'Region', 'Profit', 'Profit_zscore']])
```

```python
# Convert Order Date to datetime
dataset['Order Date'] = pd.to_datetime(dataset['Order Date'])

# Extract year and month for trend analysis
dataset['Year'] = dataset['Order Date'].dt.year
dataset['Month'] = dataset['Order Date'].dt.to_period('M')
print()
```

```python
# Sales and profit by region
region_sales_profit = dataset.groupby('Region')[['Sales', 'Profit']].sum().reset_index()
plt.figure(figsize=(10,6))
region_sales_profit_melted = region_sales_profit.melt(id_vars='Region', value_vars=['Sales', 'Profit'])
sns.barplot(data=region_sales_profit_melted, x='Region', y='value', hue='variable')
plt.title('Sales and Profit by Region')
plt.show()
```
